## 1.	Write a Python Program to Find the Factorial of a Number?

```python
# To take input from the user
num = int(input("Enter a number: "))

factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,num + 1):
       factorial = factorial*i
   print("The factorial of",num,"is",factorial)
```

## 2.	Write a Python Program to Display the multiplication Table?

```python
number = int(input ("Enter the number of which the user wants to print the multiplication table: "))      
# We are using "for loop" to iterate the multiplication 10 times       
print ("The Multiplication Table of: ", number)    
for count in range(1, 11):
  print (number, 'x', count, '=', number * count)
```

## 3.	Write a Python Program to Print the Fibonacci sequence?

```python
n_terms = int(input ("How many terms the user wants to print? "))

# First two terms  
first = 0  
second = 1  
count = 0  

# Now, we will check if the number of terms is valid or not  
if n_terms <= 0:
  print ("Please enter a positive integer, the given number is not valid")  
# if there is only one term, it will return first
elif n_terms == 1:
  print ("The Fibonacci sequence of the numbers up to", n_terms, ": ")
  print(first)  
# Then we will generate Fibonacci sequence of number  
else:
  print ("The fibonacci sequence of the numbers is:")
  while count < n_terms:
    print(first)  
    next = first + second  
# At last, we will update values  
    first = second  
    second = next
    count += 11  


```

## 4.	Write a Python Program to Check Armstrong Number?

```python
#A number is an Armstrong number if it is equal to the sum of its own digits raised to the number of digitsâ€™ power.

#enter user input
number_to_check = int(input("Enter a number to check if it is armstrong: "))

number1=list(map(int,str(number_to_check)))
number2=list(map(lambda x:x**3,number1))
if(sum(number2)==number_to_check):
	print("The given number {0} is an armstrong number. ".format(number_to_check))
else:
	print("The given number {0} isn't an arsmtrong number. ".format(number_to_check))
```

## 5.	Write a Python Program to Find Armstrong Number in an Interval?

```python
lower = int(input("Enter lower range: "))  
upper = int(input("Enter upper range: "))  

for num in range(lower,upper + 1):
  sum = 0
  temp = num  
  while temp > 0:
    digit = temp % 10 
    sum += digit ** 3  
    temp //= 10  
    if num == sum:  
      print(num)  
```

## 6.	Write a Python Program to Find the Sum of Natural Numbers?

```python
# Sum of natural numbers up to num

num = int(input("Enter a number to get sum upto: "))

if num < 0:
   print("Enter a positive number")
else:
   sum = 0
   # use while loop to iterate until zero
   while(num > 0):
       sum += num
       num -= 1
   print("The sum is", sum)
```
