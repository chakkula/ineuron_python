## 7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...

```python
def gen_primes(n):
  for number in range(1,n+1):
    if number>1:
        for i in range(2,number):
            if (number%i)==0:
                break
        else:
            yield number
```

```python
first = gen_primes(1000)
print(next(first))
```

```python
print(next(first))
print(next(first))
print(next(first))
print(next(first))
```
