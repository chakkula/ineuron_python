## 1. What is the result of the code, and why?

```python
def func(a, b=6, c=8):
  print(a, b, c)
```

```python
func(1, 2)
```

The function has three arguments that are declared with default values for 2nd and 3rd arguments to be 6 and 8. This means that the function passed with 1 argument needs a value mandatorily but for 2nd and 3rd it considers the value if passed else consider the default value to be the argument that is declared.

In the above function the 2nd argument passed has the value 2 so the function considered the vaue 2 for the argument replacing the declared value 6. Since no argument is passed for 3rd it has considered the default value 8 and printed the same.

## 3. How about this code: what is its result, and why?

```python
def func(a, *pargs):
  print(a, pargs)
```

```python
func(1, 2, 3)
```

The special syntax *args in function definitions in python is used to pass a variable number of arguments to a function. It is used to pass a non-key worded, variable-length argument list. 

## 4. What does this code print, and why?

```python
def func(a, **kargs):
  print(a, kargs)
```

The special syntax **kwargs in function definitions in python is used to pass a keyworded, variable-length argument list. We use the name kwargs with the double star. The reason is because the double star allows us to pass through keyword arguments

```python
func(a=1, c=3, b=2)
```

One can think of the kwargs as being a dictionary that maps each keyword to the value that we pass alongside it. That is why when we iterate over the kwargs there doesnâ€™t seem to be any order in which they were printed out.
