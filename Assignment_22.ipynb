## 4. What output does this code produce? Why?

```python
X = 'iNeuron'
def func():
  global X
  X = 'NI'


func()
print(X)
```

The function has called the global variable X. The value for X was 'iNeuron' but is assigned to value 'NI' inside the function. So the print statement after the func() call has returned the value for X that has changed inside the func()

## 6. How about this code: what is its output in Python 3, and explain?

```python
def func():
  X = 'NI'

def nested():
  nonlocal X
  X = 'Spam'

nested()
print(X)
func()
```

If we use nonlocal, that means that Python will, at the start of the function, look for a variable with the same name from one scope above (and beyond). But here we did not define such one.
