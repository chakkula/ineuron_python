## 1. In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

PyPDF2 Module supports PdfFileReader() and PdfFileWriter()



```python
import PyPDF2
```

```python
file = '/content/drive/MyDrive/Test_notebooks/ineuron_python/Assignments/Assignment_for_pdf.pdf'
file2 = '/content/drive/MyDrive/Test_notebooks/ineuron_python/Assignments/Assignment_for_pdf_repeat.pdf'
output = '/content/drive/MyDrive/Test_notebooks/ineuron_python/Assignments/combined.pdf'
rotate_page = '/content/drive/MyDrive/Test_notebooks/ineuron_python/Assignments/rotated_page.pdf'
```

```python
pdfFileObj = open(file, 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
pdf2FileObj = open(file2, 'rb')
pdfReader1 = PyPDF2.PdfFileReader(pdf2FileObj)
pdfWriter = PyPDF2.PdfFileWriter()
```

```python
for pageNum in range(pdfReader.numPages):
  pageObj = pdfReader.getPage(pageNum)
  pdfWriter.addPage(pageObj)
```

```python
for pageNum in range(pdfReader1.numPages):
  pageObj = pdfReader1.getPage(pageNum)
  pdfWriter.addPage(pageObj)
```

```python
pdfOutputFile = open(output, 'wb')
pdfWriter.write(pdfOutputFile)
pdfOutputFile.close()
pdfFileObj.close()
pdf2FileObj.close()
```

## 9. What is the difference between False, True, and None for the bold variable?

Runs can be further styled using text attributes. Each attribute can be set to one of three values: 

True (the attribute is always enabled, no matter what other styles are applied to the run)

False (the attribute is always disabled)

None (defaults to whatever the run’s style is set to).

## 12. What integers represent the levels of headings available in Word documents?

The arguments to add_heading() are a string of the heading text and an integer from 0 to 4. 

The integer 0 makes the heading the Title style, which is used for the top of the document. 

Integers 1 to 4 are for various heading levels, with 1 being the main heading and 4 the lowest subheading. 

The add_heading() function returns a Paragraph object to save you the step of extracting it from the Document object as a separate step.

```python
doc.add_heading('Header 0', 0)
doc.add_heading('Header 1', 1)
doc.add_heading('Header 2', 2)
doc.add_heading('Header 3', 3)
```
