## 3.	Write a Python Program to Check Leap Year?

```python
#Checking if the given year is leap year 
def CheckLeap(Year):
  if((Year % 400 == 0) or (Year % 100 != 0) and  (Year % 4 == 0)): 
    print("Year {0} is a leap Year".format(Year))
  else:
    print("Year {0} is not a leap Year".format(Year))
# Taking an input year from user  
Year = int(input("Enter the Year to check if it is a leap year?: "))  
CheckLeap(Year)
```

## 4.	Write a Python Program to Check Prime Number?

```python
def checkPrime(num):
  # define a flag variable
  flag = False
  if num > 1:
    # check for factors
    for i in range(2, num):
      if (num % i) == 0:
      # if factor is found, set flag to True
       flag = True
       # break out of loop
       break
  return flag

# To take input from the user
num = int(input("Enter a number to check for prime: "))
get_flag = checkPrime(num)

# check if flag is True
if get_flag:
    print(num, "is not a prime number")
else:
    print(num, "is a prime number")
```

## 5.	Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
lower = 1
upper = 100 #change the limit

print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```
